```python
import networkx as nx

g= nx.karate_club_graph()

# merkezilik olcutleri 

degre_merkezilik = nx.degree_centrality(g)
closense_merkezilik = nx.closeness_centrality(g)
betweenness_merkezilik = nx.betweenness_centrality(g)
eigenvector_merkezilik = nx.eigenvector_centrality(g)

def print_sorted(name,d,top=None):
   sorted_list = sorted(d.items(), key=lambda x: x[1], reverse=True )

   if top:
      sorted_list = sorted_list[:top]
   print(f"\n{name} (top {top}):")
   for n,v in sorted_list:
        print(f"Node {n}: {v:.4f}")
    

print_sorted("Degree Centrality",degre_merkezilik,top=10)
print_sorted("Closeness Centrality",closense_merkezilik,top=10)

```
