```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# okuma islemi
df_nodes = pd.read_csv("InputFileNodes.csv")
df_edges = pd.read_csv("InputFileEdges.csv")

# ıd sutunları alınıyor 
nodes=df_nodes['id'].tolist()
# from ve to sutunları alınıyor
edges_data=df_edges[['from','to']].values.tolist()
```

```python
# yönsüz graf olusturma burada yapılıyor  
G=nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges_data)

# graf çizdirme işlemi
plt.figure(figsize=(10,8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Yösüz Graf")
plt.show()
```

```python
G_directed=nx.DiGraph()
G_directed.add_nodes_from(nodes)
G_directed.add_edges_from(edges_data)

plt.figure(figsize=(10,8))
nx.draw(G_directed, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10, arrows=True)
plt.title("Yönlü Graf")
plt.show()
```

```python
# G network için tüm dğüğümlerin dereceleri silinecek 
degress= dict(G.degree())
print("Orjinal Dereceler:", degress)
```

```python
# G network için ortalama derece hesaplama 
total_degree = sum(degress.values())
num_nodes = G.number_of_nodes()
average_degree = total_degree / num_nodes
print("Ortalama Derece:", average_degree)
```

```python
# G için histogram çizme 
degree_values = list(degress.values())
max_degree = max(degree_values)
min_degree = min(degree_values)

plt.figure(figsize=(10,6))
plt.hist(degree_values, bins=range(min_degree, max_degree + 2), color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Düğüm Derece Dağılımı")
plt.xlabel("Derece")
plt.ylabel("Düğüm Sayısı")
plt.xticks(range(min_degree, max_degree + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# degree centrality hesaplama
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)
```

```python
# closeness centrality hesaplama
closeness_centrality = nx.closeness_centrality(G)
print("Closeness Centrality:", closeness_centrality)
```

```python
# betweenness centrality hesaplama
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_centrality)
```

```python
# eigenvector centrality hesaplama
eigenvector_centrality = nx.eigenvector_centrality(G)
print("Eigenvector Centrality:", eigenvector_centrality)
```

```python
# Katz centrality hesaplama
katz_centrality = nx.katz_centrality(G, alpha=0.1, beta=1.0)
print("Katz Centrality:", katz_centrality)
```

```python
# page rank hesaplama
page_rank = nx.pagerank(G, alpha=0.85)  
print("Page Rank:", page_rank)
```

```python
# modularity hesaplama
from networkx.algorithms import community
communities = community.greedy_modularity_communities(G)
modularity = community.modularity(G, communities)
print("Modularity:", modularity)
```

```python
# density hesaplama
density = nx.density(G)
print("Density:", density)
```

```python
# Average Clustering Coefficient hesaplama
avg_clustering_coefficient = nx.average_clustering(G)
print("Average Clustering Coefficient:", avg_clustering_coefficient)
```

```python
# Diameter hesaplama
if nx.is_connected(G):
    diameter = nx.diameter(G)
    print("Diameter:", diameter)    
else:
    largest_cc = max(nx.connected_components(G), key=len)
    subgraph = G.subgraph(largest_cc)
    diameter = nx.diameter(subgraph)
    print("Graf bağlı değil, çap hesaplanamaz.")
```

```python
# 19. Derece dağılımı hesaplanacak ve çizdirilecek
degree_count = nx.degree_histogram(G)
degree_prob = [d / num_nodes for d in degree_count]  # Olasılık

# Derece değerleri
degrees_k = range(len(degree_prob))

# Sadece > 0 olasılığa sahip dereceleri al
positive_prob_degrees = [(k, p) for k, p in zip(degrees_k, degree_prob) if p > 0]
k_values = [item[0] for item in positive_prob_degrees]
p_k_values = [item[1] for item in positive_prob_degrees]

plt.figure(figsize=(8, 6))

# Log-log plot (Ölçek-bağımsız ağlar için düz çizgi beklenir)
plt.loglog(k_values, p_k_values, 'o', color='red', markersize=5)
plt.title("Derece Dağılımı (Log-Log Plot)")
plt.xlabel("Derece k (log)")
plt.ylabel("P(k) (log)")
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
sorted_betweenness = sorted(betweenness_centrality.items(), key=lambda item: item[1], reverse=True)

print("21- En Kritik 5 'Köprü' Düğümü (Aradalık Merkeziliği):")
for node, value in sorted_betweenness[:5]:
    print(f"Düğüm: {node}, Değer: {value:.4f}")
```

```python
sorted_eigenvector = sorted(eigenvector_centrality.items(), key=lambda item: item[1], reverse=True)

print("\n22- En Kritik 5 'Lider' Düğümü (Özvektör Merkeziliği):")
for node, value in sorted_eigenvector[:5]:
    print(f"Düğüm: {node}, Değer: {value:.4f}")
```

```python
in_degrees = dict(G_directed.in_degree())
in_degree_values = in_degrees.values()

# dışarı derece hesaplama
out_degrees = dict(G_directed.out_degree())
out_degree_values = out_degrees.values()

# her iki dağılım için histogram çizdirme buradan yapılıyor 
plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)
max_in = max(in_degree_values) if in_degree_values else 1
min_in = min(in_degree_values) if in_degree_values else 0
plt.hist(in_degree_values, bins=range(min_in, max_in + 2), align='left', rwidth=0.8, color='darkgreen')
plt.title("Yönlü Ağ ($G_{directed}$) In-Degree Dağılımı")
plt.xlabel("In-Degree (k_in)")
plt.ylabel("Düğüm Sayısı (N)")
plt.xticks(range(min_in, max_in + 2))
plt.grid(axis='y', alpha=0.5)

plt.subplot(1, 2, 2)
max_out = max(out_degree_values) if out_degree_values else 1
min_out = min(out_degree_values) if out_degree_values else 0
plt.hist(out_degree_values, bins=range(min_out, max_out + 2), align='left', rwidth=0.8, color='darkred')
plt.title("Yönlü Ağ ($G_{directed}$) Out-Degree Dağılımı")
plt.xlabel("Out-Degree (k_out)")
plt.ylabel("Düğüm Sayısı (N)")
plt.xticks(range(min_out, max_out + 2))
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
