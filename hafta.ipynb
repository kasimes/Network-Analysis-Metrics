```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.karate_club_graph()
pos = nx.spring_layout(g)   
nx.draw(g, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
plt.show()
```

```python
print(g.nodes())
```

```python
print(g.degree([0]))
```

```python
for node,degre in g.degree():
    print(f"end={node} derece={degre}")
```

```python
hist = nx.degree_histogram(g)
plt.bar(range(len(hist)), hist) 
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution')
plt.show()
```

```python
density = nx.density(g)
print(f"Graph Density: {density}")

# agdaki mevcuk kenarlarn olas maksimum kenarlara oran olarak tanmlanr.
# Younluk, 0 ile 1 arasnda bir deer alr; 1 deeri tam bal bir grafii ifade eder.

```

```python
print (g.number_of_nodes())
# Grafikteki dm saysn dndrr
print (g.number_of_edges())
# Grafikteki kenar saysn dndrr
```

```python
nei=list(g.neighbors(0))
print(nei)
```

```python
neibords = list(nx.all_neighbors(g,0))
print(neibords)
# 0 dmnn tm komularn dndrr
```

```python
common = list(nx.common_neighbors(g, 0, 1))
print(common)
# 0 ve 1 dmleri arasndaki ortak komular dndrr
```

```python
nodes = [0, 1, 2,4,5,6,7,8,9]
subgraph = g.subgraph(nodes)
nx.draw(subgraph, with_labels=True, node_color='lightgreen', edge_color='black', node_size=500)
plt.show()
```

```python
edges=[(0, 1), (0, 2), (1, 2), (1, 3)]
subgraph_edges = g.edge_subgraph(edges)
nx.draw(subgraph_edges, with_labels=True, node_color='lightcoral', edge_color='black', node_size=500)
plt.show()
```
