2.1 Star (Yıldız) Topolojisi

Tek merkezi cihaz vardır (hub/switch).

Tüm cihazlar merkeze bağlanır.

Avantaj: Tek cihaz arızası dışında ağ çalışmaya devam eder.

Dezavantaj: Merkezi cihaz arızalanırsa tüm ağ etkilenir.

2.2 Ring (Halka) Topolojisi

Cihazlar halka şeklinde birbirine bağlanır.

Veri bir yönde veya çift yönlü halkada dolaşır.

Avantaj: Trafik dengelidir.

Dezavantaj: Tek bağlantı koparsa ağ bölünebilir.

2.3 Mesh (Örgü) Topolojisi

Her cihaz diğerleriyle doğrudan bağlantılıdır.

Avantaj: Yüksek güvenilirlik ve yedekli bağlantı.

Dezavantaj: Çok kablo ve maliyetlidir.

2.4 Bus (Omurga) Topolojisi

Tüm cihazlar tek kabloya bağlanır.

Avantaj: Basit ve ucuzdur.

Dezavantaj: Tek kablo arızası tüm ağı etkiler.

2.5 Tree (Ağaç) Topolojisi

Hiyerarşik yapı, yıldız topolojilerinin birleşimi gibi.

Büyük ağlarda yaygın.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

# 1️ Star Topology burada görselleştirilir 

G_star = nx.Graph()
nodes_star = ["Hub", "PC1", "PC2", "PC3", "PC4"]
G_star.add_nodes_from(nodes_star)
edges_star = [("Hub", "PC1"), ("Hub", "PC2"), ("Hub", "PC3"), ("Hub", "PC4")]
G_star.add_edges_from(edges_star)
nx.draw(G_star, with_labels=True, node_color='skyblue', node_size=1500)
plt.title("Star Topology")
plt.show()
```

```python
# 2️⃣ Ring Topology

G_ring = nx.cycle_graph(5)  # 5 düğüm seçildi 
labels_ring = {i:f"PC{i+1}" for i in range(5)}
nx.draw(G_ring, labels=labels_ring, with_labels=True, node_color='lightgreen', node_size=1500)
plt.title("Ring Topology")
plt.show()
```

```python


# 3️ Mesh Topology örnegi 

G_mesh = nx.complete_graph(4)  # 4 düğüm, tüm bağlantılar
labels_mesh = {i:f"PC{i+1}" for i in range(4)}
nx.draw(G_mesh, labels=labels_mesh, with_labels=True, node_color='orange', node_size=1500)
plt.title("Mesh Topology")
plt.show()
```

```python


# 4️ Bus Topology örnegii 

G_bus = nx.Graph()
nodes_bus = ["PC1", "PC2", "PC3", "PC4", "Bus"]
G_bus.add_nodes_from(nodes_bus)
edges_bus = [("PC1", "Bus"), ("PC2", "Bus"), ("PC3", "Bus"), ("PC4", "Bus")]
G_bus.add_edges_from(edges_bus)
nx.draw(G_bus, with_labels=True, node_color='pink', node_size=1500)
plt.title("Bus Topology")
plt.show()
```

```python

# 5️ Tree Topology örnegi 

G_tree = nx.Graph()
nodes_tree = ["Root", "PC1", "PC2", "PC3", "PC4"]
G_tree.add_nodes_from(nodes_tree)
edges_tree = [("Root", "PC1"), ("Root", "PC2"), ("PC2", "PC3"), ("PC2", "PC4")]
G_tree.add_edges_from(edges_tree)
nx.draw(G_tree, with_labels=True, node_color='lightyellow', node_size=1500)
plt.title("Tree Topology")
plt.show()
```
